```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from random import randint
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from IPython.display import Image
from sklearn.tree import export_graphviz
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.svm import SVR
from sklearn.svm import LinearSVR
from sklearn.utils import resample

# set to show all columns available in dataset
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
accident_data = pd.read_csv("accidents_make_sense.csv")
accident_data.head(10)
```

```python
resamplingDs = resample(accident_data, n_samples=500000, random_state=1, replace=False)
scaler = StandardScaler()
scaler.fit(resamplingDs.drop(['Accident_Severity_Percent'], axis=1))
```

```python
result=[]
result_test = []

result_train = []
result_model = []
for i in range(99, 1, -1):
    # train_test_split(feature, target, ...)
    X_train, X_test, Y_train, Y_test = train_test_split(scaler.transform(resamplingDs.drop(['Accident_Severity_Percent'], axis=1)), 
                                                        resamplingDs[['Accident_Severity_Percent']].values, 
                                                        train_size=i / 100.0, test_size=(100 - i) / 100.0, random_state=1)

    lin = LinearRegression()

    # let machine learn the training data
    lin.fit(X_train, Y_train)

    # predict against train data
    Y_predict = lin.predict(X_train)

    # calculate RMSE
    RMSE_train = np.sqrt(mean_squared_error(Y_train, Y_predict))

    # predict against test data
    Y_predict = lin.predict(X_test)

    # calculate RMSE
    RMSE_test = np.sqrt(mean_squared_error(Y_test, Y_predict))
    result.append([RMSE_train, RMSE_test])
```
